```python
from IPython.display import clear_output
import requests
from bs4 import BeautifulSoup
import string
from textblob import TextBlob
```

```python
STC = 0
total_STC = 0
total_Miles = 0

sentiment_STC = 0
doc_STC = 0
sentiment_Miles = 0
doc_Miles = 0

site = requests.get('https://www.savethechildren.org/us/about-us/media-and-news/2017-press-releases')
soup = BeautifulSoup(site.text, 'html.parser')

for div in soup.findAll('div', {'class': 'news__headline'}):
    for link in div.findAll('a'):
        href = 'https://www.savethechildren.org/' + link.get('href')
        print(href)
        
        page = requests.get(href)
        page_soup = BeautifulSoup(page.text, 'html.parser')
        
        text = page_soup.findAll('div', {'class': 'column'})[1].text
        text1 = TextBlob(text)
        
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = text.split()

        total_STC += len(text)
        total_Miles += text.count('Miles')
        STC += 1
        
        if text.count('Miles') > 0:
            sentiment_Miles += text1.sentiment.polarity
            doc_Miles += 1
        else:
            sentiment_STC += text1.sentiment.polarity
            doc_STC += 1
        
print('STC: {}, total_STC: {}, total_Miles: {}'.format(STC, total_STC, total_Miles))
print('sentiment_STC: {}, sentiment_Miles: {}, doc_Miles: {}'.format((sentiment_STC/doc_STC), (sentiment_Miles/doc_Miles), doc_Miles))



```

```python
Amnesty = 0
total_Amnesty = 0
total_Huang = 0

sentiment_Huang = 0
doc_Huang = 0
sentiment_Amnesty = 0
doc_Amnesty = 0

for i in range(82, 123):
    print('https://www.amnestyusa.org/news/newsroom/?fwp_paged=' + str(i) + '&fwp_sort=date_desc')
    site = requests.get('https://www.amnestyusa.org/news/newsroom/?fwp_paged=' + str(i) + '&fwp_sort=date_desc')
    soup = BeautifulSoup(site.text, 'html.parser')

    for link in soup.findAll('a', {'class': 'row card-group d-flex mb-32'}):

        href = 'https://www.amnestyusa.org' + link['href']
        print(href)

        page = requests.get(href)
        page_soup = BeautifulSoup(page.text, 'html.parser')
        
        try:
            text = page_soup.find('div', {'class': 'col-lg-8 mt-72'}).text
            text1 = TextBlob(text)
            
            text = text.translate(str.maketrans('', '', string.punctuation))
            text = text.split()
            total_Huang += text.count('Huang') 
            total_Amnesty += len(text)
            
            if text.count('Huang') > 0:
                sentiment_Huang += text1.sentiment.polarity
                doc_Huang += 1
            else:
                sentiment_Amnesty += text1.sentiment.polarity
                doc_Amnesty += 1
            
        except AttributeError:
            text = page_soup.find('div', {'class': 'col-lg-9 article-start'}).text    
#             print(text)
            text = text.translate(str.maketrans('', '', string.punctuation))
            text = text.split()
        
            total_Huang += text.count('Huang') 
            total_Amnesty += len(text)
            Amnesty += 1
            
print('Amnesty: {}, total_Amnesty: {}, total_Huang: {}'.format(Amnesty, total_Amnesty, total_Huang))
print('sentiment_Amnesty: {}, sentiment_Huang: {}, doc_Huang: {}'.format((sentiment_Amnesty/doc_Amnesty), (sentiment_Huang/doc_Huang), doc_Huang))

```

```python
MSF = 0
total_MSF = 0
total_Cone = 0

sentiment_Cone = 0
doc_Cone = 0
sentiment_MSF = 0
doc_MSF = 0

for i in range(32, 54):
    print('https://www.doctorswithoutborders.org/what-we-do/press-room?search_api_fulltext=&sort_by=search_api_relevance&page=' + str(i))
    site = requests.get('https://www.doctorswithoutborders.org/what-we-do/press-room?search_api_fulltext=&sort_by=search_api_relevance&page=' + str(i))
    soup = BeautifulSoup(site.text, 'html.parser')
#     print(soup)

    for div in soup.findAll('div', {'class': 'search-result search-result--node search-result--News'}):
        for link in div.findAll('a'):
            href = 'https://www.doctorswithoutborders.org' + link.get('href')
            print(href)

            page = requests.get(href)
            page_soup = BeautifulSoup(page.text, 'html.parser')

            text = page_soup.find('div', {'class': 'field__item'}).text
            text1 = TextBlob(text)

            text = text.translate(str.maketrans('', '', string.punctuation))
            text = text.split()
            
            total_MSF += len(text)
            total_Cone += text.count('Cone')
            MSF += 1
            
            if text.count('Cone') > 0:
                sentiment_Cone += text1.sentiment.polarity
                doc_Cone += 1
            else:
                sentiment_MSF += text1.sentiment.polarity
                doc_MSF += 1
        
print('MSF: {}, total_MSF: {}, total_Cone: {}'.format(MSF, total_MSF, total_Cone))
print('sentiment_MSF: {}, sentiment_Cone: {}, doc_Cone: {}'.format((sentiment_MSF/doc_MSF), 
                                                                        (sentiment_Cone/doc_Cone), doc_Cone))

```

```python
oxfam = 0
total_oxfam = 0
total_MAXMAN = 0

sentiment_Maxman = 0
doc_Maxman = 0
sentiment_Oxfam = 0
doc_Oxfam = 0

f = open("/Users/Friso/Desktop/oxfam.txt", "r")

soup = BeautifulSoup(f.read(), 'html.parser')

for dt in soup.find_all("dt"):
    div = dt.find('a')
    href = div.get('href')
    print(href)
    
    page = requests.get(href)
    page_soup = BeautifulSoup(page.text, 'html.parser')

    year = page_soup.find('div', {'class': 'cmpnt-full-article-time-and-author'})
    year = year.find('time').text[-4:]
    
    if year == '2017':
        text = page_soup.find('div', {'class': 'rich-text'}).text
        text1 = TextBlob(text)

        print('2017')
        
        total_oxfam += len(text)
        total_MAXMAN += text.count('Maxman')
        oxfam += 1
        
        if text.count('Maxman') > 0:
            sentiment_Maxman += text1.sentiment.polarity
            doc_Maxman += 1
        else:
            sentiment_Oxfam += text1.sentiment.polarity
            doc_Oxfam += 1
        
    if year == '2016':
        break
        
print('oxfam: {}, total_oxfam: {}, total_Maxman: {}'.format(oxfam, total_oxfam, total_MAXMAN))   
print('sentiment_Oxfam: {}, sentiment_Maxman: {}, doc_Maxman: {}'.format((sentiment_Oxfam/doc_Oxfam), 
                                                                        (sentiment_Maxman/doc_Maxman), doc_Maxman))
```
